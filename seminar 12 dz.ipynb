```python
import pandas as pd
from scipy import stats

df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar12/colon_cancer_tumor_vs_normal_paired_FPKM.tsv', sep='\t', index_col=0)

df
```

```python
df['unpair.test'] = stats.ttest_ind(df.iloc[:, 0:5], df.iloc[:, 5:10], axis=1)[1]
df['pair.test'] = stats.ttest_rel(df.iloc[:, 0:5], df.iloc[:, 5:10], axis=1)[1]

unpair_test = df['unpair.test']
pair_test = df['pair.test']

pair_test = pair_test[pair_test < 0.05].sort_values()
pair_test = pair_test.iloc[0:10]
print(pair_test)
```

```python
unpair_test = unpair_test[unpair_test < 0.05].sort_values()
unpair_test = unpair_test.iloc[0:10]
print(unpair_test)
```

```python
from scipy.stats import shapiro

df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar11/breast_cancer_1000_genes.tsv', sep = '\t', index_col = 0)

df
```

```python
total_genes = len(df.index)
normal = []

for gene in df.index:
    if shapiro(df.loc[gene])[1] > 0.05: #вызываем строки каждого гена по очереди
        normal.append(gene)

print(round(len(normal)/total_genes * 100, 3))
```
